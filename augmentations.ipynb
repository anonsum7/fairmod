```python
# word level augmentations
import nlpaug.augmenter.word as naw
waug_1 = naw.BackTranslationAug(device='cuda', from_model_name='facebook/wmt19-en-de', to_model_name='facebook/wmt19-de-en')
```

```python
import pickle
from tqdm import tqdm

with open('./data/comments.pkl', 'rb') as handle:
    comments = pickle.load(handle)

comments = list(set(comments))

augment_map = {}

for comment in tqdm(comments):
    augment_map[comment] = waug_1.augment(comment)
```
